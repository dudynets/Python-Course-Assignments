```python
import pandas as pd
```

```python
results = pd.read_csv('results.csv')
runners = pd.read_csv('runners.csv')

data = results.merge(runners)
data['result_average_speed'] = data['result_distance'] / data['result_time']

data
```

```python
top_3_runners = data.sort_values(['result_id', 'result_time'], ascending=False).groupby('result_id').head(3)[
    ['result_id', 'runner_name', 'runner_surname']].sort_values('result_id').reset_index(drop=True).groupby(
    'result_id').apply(lambda x: x)
top_3_runners
```

```python
number_of_rounds = data.sort_values(['result_id'], ascending=False)['result_id'].iloc[0]

finalists = data[data['result_id'] == number_of_rounds].reset_index(drop=True)[['runner_name', 'runner_surname']]
finalists
```

```python
runner_best_round = \
data.sort_values(['runner_id', 'result_average_speed'], ascending=False).groupby('runner_id').head(1).reset_index()[
    ['runner_name', 'runner_surname', 'result_id', 'result_average_speed']].sort_values('result_average_speed',
                                                                                        ascending=False)
runner_best_round
```
