```python
import pandas as pd
import ipywidgets as widgets
```

```python
clients = pd.read_csv('clients.csv')
orders = pd.read_csv('orders.csv')

data = orders.merge(clients)
data['client_full_name'] = data['client_name'] + ' ' + data['client_surname']
data['order_price'] = data['product_price'] * data['product_quantity']

data
```

```python
def filter_by_client(full_name):
    client_orders = data[data['client_full_name'] == full_name]
    client_orders = client_orders.groupby('product_name').apply(lambda product: pd.Series({
        'product_quantity': product['product_name'].count(),
    })).reset_index()

    return client_orders

widgets.interact_manual(filter_by_client, full_name=widgets.Dropdown(options=data['client_full_name'].unique(), description='Client name: '))
```

```python
clients_rating = data.groupby(['client_name', 'client_surname']).apply(lambda client_orders: pd.Series({
    'total_spent': client_orders['order_price'].sum()
})).sort_values('total_spent', ascending=False).reset_index()

clients_rating
```

```python
each_product_quantity = data.groupby('product_name').agg(quantity=('product_quantity', 'sum')).sort_values('quantity', ascending=False).reset_index()
each_product_quantity
```
